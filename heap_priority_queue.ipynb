```python
#Heaps

import heapq

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
heapq.heapify(A)


print(A)
```

```python
least=heapq.heappop(A)

print(A,least)
```

```python
heapq.heappush(A,-5)
print(A)
```

```python
def heap_sort(A):
    n=len(A)
    heapq.heapify(A)
    new_arr=[0]*n
    for i in range(n):
        new_arr[i]=heapq.heappop(A)
    
    return new_arr

print(A)
a=heap_sort(A)
print(a)
```

```python
class Node:
    def __init__(self,val,next=None,prev=None):
        self.val=val
        self.prev=prev
        self.next=next
        
    def __str__(self):
        return str(self.val)
        

A=Node(5)
B=Node(4)
C=Node(3)
D=Node(2)
E=Node(1)
     
A.next=B

B.next=C
B.prev=A

C.next=D
C.prev=B

D.next=E
D.prev=C

E.prev=D

print(A)

   
        
```

```python
def Traverse(node):
    if not node:
        return
   
    print(node,end='<->')
    Traverse(node.next)
    
    

Traverse(A)
print(None)
    
    
        
        
```

```python
def insert_node(node,head,index):
    if index == 0 and head is not None:
        node.next=head
        head.prev=node
        
        return node
    curr=head
    for _ in range(index-1):
        if curr.next is None:
            break
        curr=curr.next
    
   
    node.next=curr.next
  
    node.prev=curr
    if curr.next is not None:
        curr.next.prev=node
    curr.next=node
        
    
    
    return head
    
```

```python
nwh=insert_node(Node(6),A,2)
Traverse(nwh)
print(None)
```
