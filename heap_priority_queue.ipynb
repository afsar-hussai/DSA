```python
#Heaps

import heapq

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
heapq.heapify(A)


print(A)
```

```python
least=heapq.heappop(A)

print(A,least)
```

```python
heapq.heappush(A,-5)
print(A)
```

```python
def heap_sort(A):
    n=len(A)
    heapq.heapify(A)
    new_arr=[0]*n
    for i in range(n):
        new_arr[i]=heapq.heappop(A)
    
    return new_arr

print(A)
a=heap_sort(A)
print(a)
```
