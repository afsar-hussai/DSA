### Binary Tree

```python
class TreeNode:
    def __init__(self, val,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right
    
    def __str__(self):
        return str(self.val)
```

```python
#      1
#    2      3
# 4    5  10
```

```python
A=TreeNode(1)
B=TreeNode(2)
C=TreeNode(3)
D=TreeNode(4)
E=TreeNode(5)
F=TreeNode(10)

A.left=B
A.right=C
B.left=D
B.right=E
C.left=F

print(A)
```

#### Recursive 3 methods of Binary Tree DFS Traversal

```python
## 1. recursive preorder Traversal T.C=O(N), S.C=O(N)
## Node->Left->Right

def preorder_traversal(node):
    if not node:
        return
    print(node)
    preorder_traversal(node.left)
    preorder_traversal(node.right)

preorder_traversal(A)
```

```python
## 2. recursive In Order Traversal
## Left->Node->Right
def inorder_traversal(node):
    if not node:
        return
    inorder_traversal(node.left)
    print(node)
    inorder_traversal(node.right)

inorder_traversal(A)
```

```python
## 3. recursive Post order 
## Left->Right->Node

def postorder_traversal(node):
    if not node:
        return
    postorder_traversal(node.left)
    postorder_traversal(node.right)
    print(node)

postorder_traversal(A)
```

#### DFS by iterative approach which can be done using stacks.

```python
## 1. iterative preorder traversal
def preorder_traversal_iterative(node):
    
    
    stk=[A]
    while stk:
        node=stk.pop()
        print(node)
        if node.right: stk.append(node.right)
        if node.left: stk.append(node.left)
    
preorder_traversal_iterative(A)
```

```python
## 2. iterative inorder traversal
def inorder_traversal_iterative(node):
    stk=[]
    current=node
    while stk or current:
        while current:
            stk.append(current)
            current=current.left
        
        current=stk.pop()
        print(current)
        current=current.right
        
        
    
inorder_traversal_iterative(A)
```

```python
## 3. iterative postorder traversal using 2 stack method
def postorder_traversal_iterative(node):
    stk1=[]
    stk2=[]
    stk1.append(node)
    while stk1:
        node=stk1.pop()
        
        stk2.append(node)
        
        if node.left: stk1.append(node.left)
        if node.right: stk1.append(node.right)
        
    while stk2:
        node=stk2.pop()
        print(node)
        
    

postorder_traversal_iterative(A)
    
    
```

#### Binary Tree using BFS (Only way is queue)

```python
## T.C=S.C=O(N)

from collections import deque

def BFS(node):
    q=deque()
    q.append(node)
    while q:
        node=q.popleft()
        print(node)
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)

BFS(A)
```

#### searching in a DFS T.C=S.C=O(N)

```python
def search(node,target):
    if not node:
        return False
    if node.val == target: ## notice here we use .val instead of just node
        return True
    
    return search(node.left,target) or search(node.right,target)

search(A,10)
```
