```python
n = 8
A = [[0, 1], [1, 2], [0, 3], [3, 4], [3, 6], [3, 7], [4, 2], [4, 5], [5, 2]]

A
```

```python
# Array to Adjency matrix
M=[]
for _ in range(n):
    M.append([0]*n)
    
for u,v in A:
    M[u][v]=1

M
```

```python
# Array to Adjency list
from collections import defaultdict

D=defaultdict(list)

for u,v in A:
    D[u].append(v)

D
```

```python
#DFS recursion
def dfs_recursion(node):
    print(node)
    for nei_node in D[node]:
        if nei_node not in seen:
            
            seen.add(nei_node)
            dfs_recursion(nei_node)

seen=set()
source=0
seen.add(source)

dfs_recursion(source)
```

```python
# Iterative DFS
def iterative_dfs(node):
    seen=set()
    stk=[node]
    seen.add(node)
    while stk:
        curr=stk.pop()
        print(curr)
        for nei in D[curr]:
            if nei not in seen:
                seen.add(nei)
                stk.append(nei)
                

iterative_dfs(0)
```

```python
# BFS
from collections import deque

def BFS(node):
    q=deque()
    q.append(node)
    seen=set()
    seen.add(node)
    while q:
        curr=q.popleft()
        print(curr)
        for nei in D[curr]:
            if nei not in seen:
                seen.add(nei)
                q.append(nei)
    

BFS(0)
    
```

```python
## Custom graph

class Node:
    def __init__(self,val):
        self.val=val
        self.neighbour=[]
    
    def __str__(self):
        return str(self.val)
    
    def display(self):
        conn=[node.val for node in self.neighbour]
        return f"{self.val} is connected with {conn}"
    

A=Node('A')
B=Node('B')
C=Node('C')
D=Node('D')
E=Node('E')

A.neighbour.append(B)
B.neighbour.append(A)
A.neighbour.append(C)
C.neighbour.append(A)
C.neighbour.append(D)
D.neighbour.append(C)
B.neighbour.append(E)
E.neighbour.append(B)

B.display()
```
