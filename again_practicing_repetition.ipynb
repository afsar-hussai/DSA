```python
# Binary Trees
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

```python
# preorder node->left->right

def postOrder(node):
    if not node:
        return
    
    
    postOrder(node.left)
    
    postOrder(node.right)
    print(node,end='->')
    

postOrder(A)
print(None)
```

```python
def pre_order_iterative(node):
    stk=[node]
    
    while stk:
        n1=stk.pop()
        print(n1,end='->')
        if n1.right: stk.append(n1.right)
        if n1.left: stk.append(n1.left)
        
    print(None)
    
pre_order_iterative(A)
    
```

```python
# BFS
from collections import deque



def BFS(node):
    d=deque()
    d.append(node)
    while d:
        ele=d.popleft()
        print(ele,end='->')
        if ele.left: d.append(ele.left)
        if ele.right: d.append(ele.right)

BFS(A)
print(None)
    
```

```python
def search(node,target):
    if not node:
        return False
    if node.val == target:
        return True
    
    return search(node.left,target) or search(node.right,target)

print(search(A,5))
```

```python
def jump(nums):
    n=len(nums)
    max_reach=0
    for i in range(n):
        if i <= max_reach:
            max_reach=max(max_reach,i+nums[i])
            if max_reach >= n-1:
                return True
        else:
            return False 
    # return True

arr= [3,2,1,0,4]
print(jump(arr))
```

```python
def jump2(nums):
    n=len(nums)
    end=0
    far=0
    small=0
    for i in range(n-1):
        far=max(far,i+nums[i])
        if i == end:
            small+=1
            end=far
    return small

arr=[2,3,0,1,4]

print(jump2(arr))
```

```python
def jump2usingbacktrack(nums):
    n=len(nums)
    smallest=[float('inf')]
    
    def backtrack(i=0,jumps=0):
        if i == n-1:
            smallest[0]=min(smallest[0],jumps)
        
        max_reachable_index=min(i+nums[i],n-1)
        
        for new_index in range(max_reachable_index,i,-1):
            backtrack(new_index,jumps+1)
    backtrack()
    return smallest[0]


arr=[2,3,0,1,4]

print(jump2usingbacktrack(arr))
```

```python
def LPS(needle)->list[int]:
    n=len(needle)
    i=1
    prevLPS=0
    lps=[0]*n
    while i < n:
        if needle[i] == needle[prevLPS]:
            prevLPS+=1
            lps[i]=prevLPS
            i+=1
        else:
            if prevLPS!=0:
                prevLPS=lps[prevLPS-1]
            else:
                i+=1
    return lps

def KMP(haystack,needle)->int:
    n=len(haystack)
    lps=LPS(needle)
    m=len(needle)
    i=0
    j=0
    while i<n:
        
        if haystack[i] == needle[j]:
            i+=1
            j+=1
            if j == m:
                return i-m
        else:
            if j>0:
                j=lps[j-1]
            elif j == 0:
                i+=1
    return -1
        
    

haystack="sadbutsad"
needle="sad"
print(LPS(needle))
print(KMP(haystack,needle))
```

```python
words = ["This", "is", "an", "example", "of", "text", "justification."] 
maxWidth = 16

n=len(words)
i=0
ans=[]

while i < n:
    j=i+1
    
    line=""
    line_length=len(words[i])
    # Finding how many words fit in this line
    while j<n and line_length+len(words[j])+1<=maxWidth:
        line_length+=len(words[j])+1
        j+=1
    word_count=j-i
    
    
    #finding number of spaces
    
    k=i
    total_letters=0
    while k<j:
        total_letters+=len(words[k])
        k+=1
    
    spaces=maxWidth-total_letters
    
    if j ==n:
        #last line
        line+=" ".join(words[i:j])
        trailling_spaces=maxWidth-len(line)
        line+=trailling_spaces*" "
        ans.append(line)
    elif j-i == 1:
        # one word
        line+=words[i]+spaces*" "
        ans.append(line)
        
    else:
        # middle
        slots=word_count-1
        spaces_in_each_slot=spaces//slots
        remaining_spaces=spaces%slots
        k=i
        while k<j-1:
            if remaining_spaces>0:
                line+=words[k]+(spaces_in_each_slot+1)*" "
                remaining_spaces-=1
            else:
                line+=words[k]+spaces_in_each_slot*" "
            k+=1
        line+=words[j-1]
        ans.append(line)
    i=j
        
ans
```

```python
#reversing a LL
def reverse_LL(head):
    current=head
    prev=None
    next=current.next
    
    while current:
        next=current.next
        current.next=prev
        prev=current
        current=next
    
    return prev

        
    
```

```python
class ListNode:
    def __init__(self,val=0,next=None):
        self.val=val
        self.next=next

def merge(l1,l2):
    if not l1: return l2
    if not l2: return l1
    
    if l1.val<=l2.val:
        l1.next=merge(l1.next,l2)
        return l1
    else:
        l2.next=merge(l1,l2.next)
        return l2
        

def lists():
    
        
    
    A=ListNode(1)
    B=ListNode(3)
    C=ListNode(4)
    
    A1=ListNode(1)
    B1=ListNode(2)
    C1=ListNode(4)
    
    A.next=B
    B.next=C
    
    A1.next=B1
    B1.next=C1
    
    return A,A1
    
def print_lists(head):
    current=head
    
    while current:
        print(current.val,end="->")
        current=current.next
    print(None)
    
def answer():
    list1,list2=lists()
    print("Before sort")
    print_lists(list1)
    print_lists(list2)
    
    
    st=merge(list1,list2)
    
    print("After Sort11")
    print_lists(st)
    
lists()
answer()

```
