```python
# Binary Trees
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

```python
# preorder node->left->right

def postOrder(node):
    if not node:
        return
    
    
    postOrder(node.left)
    
    postOrder(node.right)
    print(node,end='->')
    

postOrder(A)
print(None)
```

```python
def pre_order_iterative(node):
    stk=[node]
    
    while stk:
        n1=stk.pop()
        print(n1,end='->')
        if n1.right: stk.append(n1.right)
        if n1.left: stk.append(n1.left)
        
    print(None)
    
pre_order_iterative(A)
    
```

```python
# BFS
from collections import deque



def BFS(node):
    d=deque()
    d.append(node)
    while d:
        ele=d.popleft()
        print(ele,end='->')
        if ele.left: d.append(ele.left)
        if ele.right: d.append(ele.right)

BFS(A)
print(None)
    
```

```python
def search(node,target):
    if not node:
        return False
    if node.val == target:
        return True
    
    return search(node.left,target) or search(node.right,target)

print(search(A,5))
```

```python
# BST
# Binary Search Trees (BSTs)

#       5
#    1    8
#  -1 3  7 9

A2 = TreeNode(5)
B2 = TreeNode(1)
C2 = TreeNode(8)
D2 = TreeNode(-1)
E2 = TreeNode(3)
F2 = TreeNode(7)
G2 = TreeNode(9)

A2.left, A2.right = B2, C2
B2.left, B2.right = D2, E2
C2.left, C2.right = F2, G2

print(A2)
```

```python
def BST(node,target):
    if not node:
        return False
    
    if  node.val == target:
        return True
    
    if target < node.val: return BST(node.left,target)
    else: return BST(node.right,target)
    
print(BST(A2,10))
```
